```python
import pandas as pd

burritos=pd.read_csv("D:/FALL_2020/CSE578_DV/Homeworks/homework-05-innovative-vis-imjaya/data/Burritos_preprocessed_new.csv")
```

```python
burritos.head()
```

```python
for i in range(len(burritos["Burrito"])):
    burritos["Burrito"][i]=burritos["Burrito"][i].lower()
```

```python
L=list(burritos['Location'])
```

```python
print(len(set(L)))
```

```python
freq = {} 
for item in L: 
    if (item in freq): 
        freq[item] += 1
    else: 
        freq[item] = 1
freq = sorted(freq.items(), key=lambda kv: kv[1],reverse=True)
```

```python
print(freq)
```

```python
for i in freq:
    if(i[1]>=2):
        print(i[0],i[1])
```

```python
address=[]
for i in range(len(burritos['Address'])):
    address.append(str(burritos['Address'][i])+","+str(burritos["Neighborhood"][i]))
    
```

```python
len(address)
```

```python
burritos['Address']=address
```

```python
burritos=burritos.drop('Neighborhood',1)
```

```python
burritos.head()
```

```python
burritos=burritos.sort_values(by=['Location'])
```

```python
len(burritos)
```

```python
burritos=burritos.reset_index(drop=True)
```

```python
burritos.head()
```

```python
burritos=burritos.fillna(0)
```

```python
burritos=burritos.drop_duplicates(subset="Location")
```
